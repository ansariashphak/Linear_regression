# Linear Regression  Concept  in Data Analysis

```python
import numpy as np

years_of_experience = np.array([1, 2, 3, 4, 5])
salary = np.array([31000, 33000, 34000, 37000, 38000])
```

```python
mean_x = np.mean(years_of_experience)
mean_y= np.mean(salary)

sum_xy = 0
sum_xx = 0

for i in range(len(years_of_experience)):
    x = years_of_experience[i]
    y = salary[i]
    
    sum_xy += (x-mean_x) * (y - mean_y)
    sum_xx += (x-mean_x) ** 2 
    
m = sum_xy / sum_xx
b = mean_y - m * mean_x

print("value of m is",m,"value of b is",b)
```

# Plot the data points/ Plot the best-fit line

```python
import matplotlib.pyplot as plt


plt.scatter(years_of_experience, salary, color='blue', label='Data points')


plt.plot(years_of_experience, m * years_of_experience + b, color='red', label='Best-fit line')


plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Salary vs Years of Experience')
plt.legend()


plt.grid(True)
plt.show()
```

# LinearRegression Model

```python
from sklearn.linear_model import LinearRegression
import numpy as np


years_of_experience = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
salary = np.array([31000, 33000, 34000, 37000, 38000])

# Create and fit the linear regression model
model = LinearRegression().fit(years_of_experience, salary)

# Retrieve the slope (m) and y-intercept (b)
m = model.coef_[0]
b = model.intercept_

print("Slope (m):", m)
print("Y-intercept (b):", b)
```

```python
import matplotlib.pyplot as plt

# Plot the data points
plt.scatter(years_of_experience, salary, color='blue', label='Data points')

# Plot the best-fit line
plt.plot(years_of_experience,model.predict(years_of_experience), color='red', label='Best-fit line')

# Add labels and legend
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Salary vs Years of Experience')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
